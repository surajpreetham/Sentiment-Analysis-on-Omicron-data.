- PorterStemmer and WordNetLemmatizer: They are used to get the root word of the word. 
- tokenization: Segmenting the words in a sentence.

###  <mark> <span  style= "font-family: Times New Roman">Preprocessing techniques </span> </mark> <a class="anchor" id="Import_necessary_libraries"></a>

````python
# Remove https links or urls
def remove_html(text):
    return re.sub(r'https://\S+','',text)

# Remove tags in the tweets
def remove_tags(text):
    return re.sub('@[A-Za-z0-9-]+([A-Z|a-z])','',text)

# stripping html texts
def strip_html(text):
    soup = BeautifulSoup(text,"html.parser") #Remove https words
    return soup.get_text()

# REmove all special characters between brackets
def remove_between_square_brackets(text):
    return re.sub('\[[^]]*\]','',text)   #Remove special characters

# Removing special characters including no's
def remove_special_characters(text,remove_digits=True):
    pattern = r'[^a-zA-Z0-9\s]'
    text = re.sub(pattern,'',text)
    return text

# Removing numbers from the data
def remove_numbers(text):
    pattern = r'\d+'
    text = re.sub(pattern,'',text)
    return text

# Remove individual letters from the data:
def remove_individual_letters(text):
    pattern = r'\ [a-z] '
    text = re.sub(pattern,'',text)
    return text

# Denoising text
def denoise_text(text):
    text = remove_html(text)
    text = strip_html(text)
    text = remove_between_square_brackets(text)
    text = remove_tags(text)
    text = remove_special_characters(text)
    text = remove_numbers(text)
    text = remove_individual_letters(text)
    return text  #Function call
  
df_new['tweets_new'] = df_new['tweets'].apply(denoise_text)
df_new.head(5)
````

## LSTM RNN

````python
from keras.models import Sequential
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
````

````python
df_lstm = pd.read_csv('twitter_data.csv')
````

````python
df_lstm.head()
````

````python
df_lstm['tweets'] = df_lstm['tweets'].apply(denoise_text)
df_lstm['tweets'] = df_lstm['tweets'].apply(remove_stopwords)
df_lstm['tweets'] = df_lstm['tweets'].apply(simpleLemmatization)
df_lstm['tweets'] = df_lstm['tweets'].apply(remove_stopwords)
df_lstm.head()
````

````python
df_lstm = df_lstm[['tweets']]
df_lstm.head(2)
````

````python
## Lets add polarity to it. 
pol = lambda x:TextBlob(x).sentiment.polarity
df_lstm['polarity'] = df_lstm['tweets'].apply(pol)
df_lstm.head()
````

````python
df_lstm['polarity'] = df_lstm['polarity'].apply(lambda x: 1 if x>0 else 0)
df_lstm.head(2)
````

````python
df_lstm['polarity'].value_counts()
````

````python
# After tokenising, num_words is a hyperparameter to 500 based on the frequency
tokenizer_new = Tokenizer(num_words=500, split=' ') 
# Fit on text --> Encoding
tokenizer_new.fit_on_texts(df_lstm['tweets'].values)
# tokenizer --> texts to sequences for tweets
X = tokenizer_new.texts_to_sequences(df_lstm['tweets'].values)
X = pad_sequences(X)
````

````python
X.shape
````

````python
## Lets build the model
model = Sequential()
# Adding a layer for words embedding 
model.add(Embedding(500,120,input_length=X.shape[1]))
# Adding dropout regularization to regulate overfitting
model.add(SpatialDropout1D(0.4))
# Add LSTM layer
model.add(LSTM(256,dropout=0.2,recurrent_dropout=0.2))
# Add a dense layer
model.add(Dense(2,activation='softmax'))
# Model compiling using entropy 
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
````

````python
# Splitting the data to train test data
from sklearn.model_selection import train_test_split
y = pd.get_dummies(df_lstm['polarity'])
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
````

````python
# Train the model
batch_size=32
model.fit(X_train, y_train, epochs = 5, batch_size=batch_size, verbose = 'auto')
````

````python
# Evaluating the model
# loss function value and accuracies are found
scores = model.evaluate(X_test,y_test)
print('The accuracy of the LSTM model is: ',scores[1])
````
